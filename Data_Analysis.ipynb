```python
job_salary = df.groupby('Job Role', as_index=False)['Salary'].mean()

top_10_job_salary = job_salary.sort_values(by='Salary', ascending=False).head(10)

plt.figure(figsize=(15, 7))
colors = sns.color_palette("Set2", len(top_10_job_salary))  # colors for bars

ax = sns.barplot(data=top_10_job_salary, x='Job Role', y='Salary', hue='Job Role', palette=colors, legend=False)

plt.xticks(rotation=45)

for bars in ax.containers:
    ax.bar_label(bars, fmt='%.0f')

plt.title("Top 10 Job Roles by Average Salary")
plt.xlabel("Job Role")
plt.ylabel("Average Salary")
plt.tight_layout()
plt.show()
```

```python
salary_role_gender = df.groupby(['Job Role', 'Gender'], as_index=False)['Salary'].mean()

count_role_gender = df.groupby(['Job Role', 'Gender']).size().reset_index(name='Count')

top_10_roles = df.groupby('Job Role')['Salary'].mean().sort_values(ascending=False).head(10).index
salary_role_gender_top10 = salary_role_gender[salary_role_gender['Job Role'].isin(top_10_roles)]
count_role_gender_top10 = count_role_gender[count_role_gender['Job Role'].isin(top_10_roles)]

plt.figure(figsize=(16, 7))
plt.subplot(1, 2, 1)
ax1 = sns.barplot(data=salary_role_gender_top10, x='Job Role', y='Salary', hue='Gender', palette='pastel')
plt.xticks(rotation=45)
plt.title('Average Salary by Job Role and Gender (Top 10 Roles)')
plt.xlabel('Job Role')
plt.ylabel('Average Salary')
for container in ax1.containers:
    ax1.bar_label(container, fmt='%.0f')

plt.subplot(1, 2, 2)
ax2 = sns.barplot(data=count_role_gender_top10, x='Job Role', y='Count', hue='Gender', palette='pastel')
plt.xticks(rotation=45)
plt.title('Employee Count by Job Role and Gender (Top 10 Roles)')
plt.xlabel('Job Role')
plt.ylabel('Count of Employees')
for container in ax2.containers:
    ax2.bar_label(container)

plt.tight_layout()
plt.show()
```

```python
top_roles = df['Job Role'].value_counts().head(10).index
filtered_df = df[df['Job Role'].isin(top_roles)]

plt.figure(figsize=(14, 6))
ax = sns.countplot(data=filtered_df, x='Job Role', hue='Department Type', palette='Set2')

for bars in ax.containers:
    ax.bar_label(bars)
    
plt.title('Top 10 Job Roles by Department Type')
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_roles = df['Job Role'].value_counts().head(23).index
filtered_df = df[df['Job Role'].isin(top_roles)]

plt.figure(figsize=(18, 9))
ax = sns.countplot(data=filtered_df, x='Job Role', hue='Gender', palette='Set2')

for bars in ax.containers:
    ax.bar_label(bars)
    
plt.title('To show Gender M & F comparison corresponding to each job role')
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
wlb_dept = df.groupby('Department Type', as_index=False)['Work-Life Balance Score'].mean()

plt.figure(figsize=(8, 5))
ax = sns.barplot(data=wlb_dept, 
                 x='Department Type', 
                 y='Work-Life Balance Score', 
                 hue='Department Type', 
                 palette='pastel',
                 legend=False) 

for bars in ax.containers:
    ax.bar_label(bars, fmt='%.2f')

# Title and axis labels
plt.title("Average Work-Life Balance Score by Department Type")
plt.xlabel("Department Type")
plt.ylabel("Average Work-Life Balance Score")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
